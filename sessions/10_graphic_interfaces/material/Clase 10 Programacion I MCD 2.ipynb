##### <b> <font color=  #6600CC > <i> Ventanas Emergentes

Ventana emergente que permite guardar un archivo

```python
from tkinter import *
from tkinter import messagebox as MessageBox
from tkinter import colorchooser as ColorChooser
from tkinter import filedialog as FileDialog

# Configuración de la raíz
raiz = Tk()


def test():
	fichero = FileDialog.asksaveasfile(title="Guardar un fichero", mode="w", defaultextension=".txt") 
	if fichero is not None:
		fichero.write("Este archivo fué creado desde Python")
		fichero.close()


Button(raiz, text="Clícame", command=test).pack()


raiz.mainloop()
```
