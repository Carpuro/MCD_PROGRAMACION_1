## Ejercicio 1

1. Utiliza el comando open para crear un archivo de texto que contenga al menos 120 caracteres distribuidos en tres líneas
2. Imprime en pantalla los caracteres que se encuentren de la posición 15 en adelante
3. Imprime en pantalla la segunda línea del texto
4. Agrega una cuarta linea al archivo
5. Imprime los primeros 5 caracteres y, sin cerrar el archivo, imprime el texto completo contenido en él
6. Cierra el archivo.

```python
# Ejercicio 1

print("="*70)
print("EJERCICIO 1: MANEJO DE ARCHIVOS DE TEXTO")
print("="*70)

# 1. Crear archivo con al menos 120 caracteres en tres líneas
print("\n1. Creando archivo de texto con 3 líneas...")

with open('archivo_ejercicio.txt', 'w') as archivo:
    linea1 = "Esta es la primera línea del archivo con suficientes caracteres para cumplir el requisito.\n"
    linea2 = "La segunda línea también contiene bastante texto para asegurar más de 120 caracteres.\n"
    linea3 = "Finalmente, la tercera línea completa el archivo con el contenido necesario para este ejercicio.\n"
    
    archivo.write(linea1)
    archivo.write(linea2)
    archivo.write(linea3)

print("✓ Archivo creado: 'archivo_ejercicio.txt'")

# Verificar longitud
texto_completo = linea1 + linea2 + linea3
print(f"✓ Total de caracteres: {len(texto_completo)}")

# 2. Imprimir caracteres desde la posición 15 en adelante
print("\n" + "-"*70)
print("2. Caracteres desde la posición 15 en adelante:")
print("-"*70)

with open('archivo_ejercicio.txt', 'r') as archivo:
    contenido = archivo.read()
    print(contenido[15:])

# 3. Imprimir la segunda línea
print("-"*70)
print("3. Segunda línea del texto:")
print("-"*70)

with open('archivo_ejercicio.txt', 'r') as archivo:
    lineas = archivo.readlines()
    print(lineas[1], end='')

# 4. Agregar una cuarta línea
print("\n" + "-"*70)
print("4. Agregando cuarta línea al archivo...")
print("-"*70)

with open('archivo_ejercicio.txt', 'a') as archivo:
    linea4 = "Esta es la cuarta línea agregada posteriormente al archivo.\n"
    archivo.write(linea4)

print("✓ Cuarta línea agregada")

# 5. Imprimir primeros 5 caracteres y luego texto completo (sin cerrar)
print("-"*70)
print("5. Primeros 5 caracteres:")
print("-"*70)

archivo = open('archivo_ejercicio.txt', 'r')
contenido_total = archivo.read()
print(contenido_total[:5])

print("\n" + "-"*70)
print("Texto completo del archivo:")
print("-"*70)
archivo.seek(0)  # Volver al inicio del archivo
print(archivo.read())

# 6. Cerrar el archivo
print("-"*70)
print("6. Cerrando el archivo...")
print("-"*70)
archivo.close()
print("Archivo cerrado")
print("-"*70)
```

## Ejercicio 2

1. Utilizando los comandos de python que acabamos de ver, crea un archivo de tipo excel que contenga información en dos hojas
2. La primer hoja se llamará "Información General"
3. La segunda se llamará "Transacciones Septiembre".

```python
# Ejercicio 2
print("="*70)
print("EJERCICIO 2: CREACIÓN DE ARCHIVO EXCEL CON MÚLTIPLES HOJAS")
print("="*70)

# 1. Crear archivo Excel con dos hojas
print("\n1. Creando archivo Excel con dos hojas...")

# Crear datos para la primera hoja: "Información General"
info_general = {
    'ID': [1, 2, 3, 4, 5],
    'Nombre': ['Juan Pérez', 'María García', 'Carlos López', 'Ana Martínez', 'Luis Rodríguez'],
    'Edad': [28, 34, 45, 29, 38],
    'Ciudad': ['Guadalajara', 'CDMX', 'Monterrey', 'Puebla', 'Querétaro'],
    'Departamento': ['Ventas', 'TI', 'Marketing', 'Recursos Humanos', 'Finanzas']
}

df_info = pd.DataFrame(info_general)

print("\n✓ Datos para 'Información General':")
print(df_info)

# Crear datos para la segunda hoja: "Transacciones Septiembre"
transacciones = {
    'Fecha': ['2024-09-01', '2024-09-05', '2024-09-10', '2024-09-15', 
              '2024-09-20', '2024-09-25', '2024-09-30'],
    'Cliente': ['Cliente A', 'Cliente B', 'Cliente C', 'Cliente A', 
                'Cliente D', 'Cliente B', 'Cliente E'],
    'Producto': ['Laptop', 'Mouse', 'Teclado', 'Monitor', 
                 'Webcam', 'Audífonos', 'Laptop'],
    'Cantidad': [2, 10, 5, 3, 8, 15, 1],
    'Precio_Unitario': [15000, 250, 800, 3500, 1200, 450, 15000],
    'Total': [30000, 2500, 4000, 10500, 9600, 6750, 15000]
}

df_transacciones = pd.DataFrame(transacciones)

print("\n✓ Datos para 'Transacciones Septiembre':")
print(df_transacciones)

# 2 y 3. Guardar en Excel con nombres específicos de hojas
print("\n" + "-"*70)
print("2 y 3. Guardando archivo Excel con hojas nombradas...")
print("-"*70)

nombre_archivo = 'datos_ejercicio.xlsx'

# Crear el archivo Excel con múltiples hojas
with pd.ExcelWriter(nombre_archivo, engine='openpyxl') as writer:
    # Hoja 1: Información General
    df_info.to_excel(writer, sheet_name='Información General', index=False)
    
    # Hoja 2: Transacciones Septiembre
    df_transacciones.to_excel(writer, sheet_name='Transacciones Septiembre', index=False)

print(f"✓ Archivo creado: '{nombre_archivo}'")
print(f"✓ Hoja 1: 'Información General'")
print(f"✓ Hoja 2: 'Transacciones Septiembre'")

# Verificar el archivo creado
print("\n" + "-"*70)
print("VERIFICACIÓN: Leyendo el archivo creado")
print("-"*70)

# Leer hoja 1
print("\nContenido de 'Información General':")
df_verificacion1 = pd.read_excel(nombre_archivo, sheet_name='Información General')
print(df_verificacion1)

# Leer hoja 2
print("\nContenido de 'Transacciones Septiembre':")
df_verificacion2 = pd.read_excel(nombre_archivo, sheet_name='Transacciones Septiembre')
print(df_verificacion2)

# Mostrar información adicional
print("\n" + "-"*70)
print("INFORMACIÓN DEL ARCHIVO:")
print("-"*70)

# Leer todas las hojas
xls = pd.ExcelFile(nombre_archivo)
print(f"Hojas en el archivo: {xls.sheet_names}")
print(f"Total de hojas: {len(xls.sheet_names)}")
```

## Ejercicio 3
1. Crea un archivo JSON llamado "Organigrama.json" en tu computadora utilizando la información contenida en el objeto "Organigrama"
2. Un vez creado, léelo y muestra la información en pantalla.

```python
# Ejercicio 3

print("="*70)
print("EJERCICIO 3: CREACIÓN Y LECTURA DE ARCHIVO JSON")
print("="*70)

# Datos del organigrama
organigrama = [
    {
        "departamento": 1,
        "nombre_depto": "Recursos Humanos",
        "director": "Laura Godinez",
        "empleados": [
            {
                "nombre": "Paul",
                "apellido": "Fierro"
            },
            {
                "nombre": "Jose",
                "apellido": "Bonilla"
            }
        ]  
    },
    {
        "departamento": 2,
        "nombre_depto": "Ingenierias",
        "director": "Salvador Medrano",
        "empleados": [
            {
                "nombre": "Carlos",
                "apellido": "Oropeza"
            },
            {
                "nombre": "Manuel",
                "apellido": "Medina"
            },
            {
                "nombre": "Roberto",
                "apellido": "Aguirre"
            }
        ]
    }   
]

# 1. Crear archivo JSON
print("\n1. Creando archivo 'Organigrama.json'...")
print("-"*70)

nombre_archivo = 'Organigrama.json'

with open(nombre_archivo, 'w', encoding='utf-8') as archivo:
    json.dump(organigrama, archivo, indent=2, ensure_ascii=False)

print(f"✓ Archivo creado: '{nombre_archivo}'")

# 2. Leer el archivo y mostrar en pantalla
print("\n2. Leyendo archivo y mostrando información...")
print("-"*70)

with open(nombre_archivo, 'r', encoding='utf-8') as archivo:
    datos_leidos = json.load(archivo)

print("\nContenido del archivo 'Organigrama.json':")
print("="*70)

# Mostrar de forma legible
for depto in datos_leidos:
    print(f"\n{'='*70}")
    print(f"DEPARTAMENTO {depto['departamento']}: {depto['nombre_depto']}")
    print(f"{'='*70}")
    print(f"Director: {depto['director']}")
    print(f"\nEmpleados:")
    print("-"*70)
    
    for i, empleado in enumerate(depto['empleados'], 1):
        print(f"  {i}. {empleado['nombre']} {empleado['apellido']}")

# Mostrar también en formato JSON
print("\n" + "="*70)
print("Formato JSON completo:")
print("="*70)
print(json.dumps(datos_leidos, indent=2, ensure_ascii=False))

# Información adicional
print("\n" + "="*70)
print("ESTADÍSTICAS:")
print("="*70)
print(f"Total de departamentos: {len(datos_leidos)}")

for depto in datos_leidos:
    num_empleados = len(depto['empleados'])
    print(f"  - {depto['nombre_depto']}: {num_empleados} empleados")

total_empleados = sum(len(d['empleados']) for d in datos_leidos)
print(f"\nTotal de empleados: {total_empleados}")
print("="*70)
```

```python
print("="*70)
print("EJERCICIO 4: CONVERSIÓN Y MANEJO DE MÚLTIPLES FORMATOS")
print("="*70)

# 1. Leer archivo Excel
print("\n1. Leyendo archivo 'Clientes.xlsx'...")
print("-"*70)

df = pd.read_excel('material/Clientes.xlsx')
print("✓ Archivo leído correctamente")

# 2. Mostrar el archivo en pantalla
print("\n2. Contenido del archivo:")
print("-"*70)
print(df)
print(f"\nDimensiones: {df.shape[0]} filas x {df.shape[1]} columnas")

# 3. Convertir a XML y guardar
print("\n3. Convirtiendo a XML y guardando como 'Ejercicio4.xml'...")
print("-"*70)

# Convertir DataFrame a XML sin índice
xml_data = df.to_xml(index=False)

with open('Ejercicio4.xml', 'w', encoding='utf-8') as f:
    f.write(xml_data)

print("✓ Archivo XML creado: 'Ejercicio4.xml'")

# 4. Crear archivos de cada tipo
print("\n" + "="*70)
print("4. CREANDO ARCHIVOS EN DIFERENTES FORMATOS")
print("="*70)

# Usar los datos de Clientes.xlsx leídos anteriormente
print("\nUsando los datos de 'Clientes.xlsx' para crear todos los formatos:")
print(df)

# 4.1 Crear archivo de TEXTO
print("\n" + "-"*70)
print("4.1 Creando archivo de TEXTO...")
print("-"*70)

with open('ejercicio4_datos.txt', 'w', encoding='utf-8') as f:
    f.write("DATOS DE CLIENTES\n")
    f.write("="*50 + "\n\n")
    
    for index, row in df.iterrows():
        for col in df.columns:
            f.write(f"{col}: {row[col]}\n")
        f.write("-"*50 + "\n")

print("✓ Archivo creado: 'ejercicio4_datos.txt'")

# 4.2 Crear archivo CSV
print("\n" + "-"*70)
print("4.2 Creando archivo CSV...")
print("-"*70)

df.to_csv('ejercicio4_datos.csv', index=False, encoding='utf-8')
print("✓ Archivo creado: 'ejercicio4_datos.csv'")

# 4.3 Crear archivo Excel
print("\n" + "-"*70)
print("4.3 Creando archivo Excel...")
print("-"*70)

df.to_excel('ejercicio4_datos.xlsx', index=False, sheet_name='Clientes')
print("✓ Archivo creado: 'ejercicio4_datos.xlsx'")

# 4.4 Crear archivo JSON
print("\n" + "-"*70)
print("4.4 Creando archivo JSON...")
print("-"*70)

datos_json = df.to_dict(orient='records')

with open('ejercicio4_datos.json', 'w', encoding='utf-8') as f:
    json.dump(datos_json, f, indent=2, ensure_ascii=False)

print("✓ Archivo creado: 'ejercicio4_datos.json'")

# 4.5 Crear archivo XML
print("\n" + "-"*70)
print("4.5 Creando archivo XML...")
print("-"*70)

xml_string = df.to_xml(index=False)

with open('ejercicio4_datos.xml', 'w', encoding='utf-8') as f:
    f.write(xml_string)

print("✓ Archivo creado: 'ejercicio4_datos.xml'")

# 5. Leer todos los archivos y mostrar información
print("\n" + "="*70)
print("5. LEYENDO TODOS LOS ARCHIVOS CREADOS")
print("="*70)

# 5.1 Leer archivo de TEXTO
print("\n" + "-"*70)
print("5.1 Leyendo archivo TEXTO:")
print("-"*70)

with open('ejercicio4_datos.txt', 'r', encoding='utf-8') as f:
    contenido_txt = f.read()
    print(contenido_txt)

# 5.2 Leer archivo CSV
print("\n" + "-"*70)
print("5.2 Leyendo archivo CSV:")
print("-"*70)

df_csv = pd.read_csv('ejercicio4_datos.csv')
print(df_csv)

# 5.3 Leer archivo Excel
print("\n" + "-"*70)
print("5.3 Leyendo archivo Excel:")
print("-"*70)

df_excel = pd.read_excel('ejercicio4_datos.xlsx')
print(df_excel)

# 5.4 Leer archivo JSON
print("\n" + "-"*70)
print("5.4 Leyendo archivo JSON:")
print("-"*70)

with open('ejercicio4_datos.json', 'r', encoding='utf-8') as f:
    datos_json_leidos = json.load(f)

print(json.dumps(datos_json_leidos, indent=2, ensure_ascii=False))

# También como DataFrame
df_json = pd.DataFrame(datos_json_leidos)
print("\nComo DataFrame:")
print(df_json)

# 5.5 Leer archivo XML
print("\n" + "-"*70)
print("5.5 Leyendo archivo XML:")
print("-"*70)

df_xml = pd.read_xml('ejercicio4_datos.xml')
print(df_xml)

# Resumen final
print("\n" + "="*70)
print("RESUMEN DE ARCHIVOS CREADOS Y LEÍDOS")
print("="*70)

archivos_creados = [
    'ejercicio4_datos.txt',
    'ejercicio4_datos.csv',
    'ejercicio4_datos.xlsx',
    'ejercicio4_datos.json',
    'ejercicio4_datos.xml',
    'Ejercicio4.xml'
]

print("\nArchivos creados:")
for i, archivo in enumerate(archivos_creados, 1):
    print(f"  {i}. {archivo}")



```
