#### <b> <font color= #990000 > <i> Creando un archivo de texto nuevo

Un archivo se puede abrir en tres modos diferentes: 

    Modo Lectura (r)
    Modo Escritura (w)
    Modo Adjuntar (a)

```python
from io import open

#Utilizamos el comando open, pasándole dos parámetros: El nombre del archivo y el modo en el que lo queremos abrir, evitar acentos en el nombre o en caso de erorr UTF-8 agregar el argumento , encoding='utf-8'
archivo_nuevo = open("archivo_nuevo.txt", "w")
```

```python
texto="Estamos creando un archivo de texto desde Jupyter Notebook saludos 22-10-25 18:16"

archivo_nuevo.write(texto)
```

```python
archivo_nuevo.close()
```

#### <b> <font color= #990000 > <i> Leyendo un archivo de texto

```python
#Utilizamos el comando open, pasándole dos parámetros: El nombre del archivo y el modo en el que lo queremos abrir 
archivo_leido = open("archivo_nuevo.txt", "r")

contenido = archivo_leido.read()

archivo_leido.close()

print(contenido)
```

#### <b> <font color= #990000 > <i> Agregando información en un archivo de texto

```python
#Utilizamos el comando open, pasándole dos parámetros: El nombre del archivo y el modo en el que lo queremos abrir 
archivo_leido = open("archivo_nuevo.txt", "a")

archivo_leido.write("\nEste es el tema número 7 de la Materia 22-oct-25 18:20")

archivo_leido.close()
```

#### <b> <font color= #990000 > <i> Obteniendo las lineas de texto

```python
#Utilizamos el comando open, pasándole dos parámetros: El nombre del archivo y el modo en el que lo queremos abrir 
archivo_leido = open("archivo_nuevo.txt", "r")

contenido = archivo_leido.readlines()

archivo_leido.close()
```

```python
print(contenido[1])
```

#### <b> <font color= #990000 > <i> Obteniendo una lista que contenga las lineas de texto

```python
#Utilizamos el comando open, pasándole dos parámetros: El nombre del archivo y el modo en el que lo queremos abrir 
archivo_leido = open("archivo_nuevo.txt", "r+") #Lectura y escritura

print(archivo_leido.readlines())

archivo_leido.close()
```

#### <b> <font color= #990000 > <i> Punteros en Texto

Si queremos que el puntero o cursor se coloque en una posición determinada dentro del archivo, tenemos que utilizar la función "seek" e indicarle la posición en la que estamos interesados

```python
#Utilizamos el comando open, pasándole dos parámetros: El nombre del archivo y el modo en el que lo queremos abrir 
archivo_leido = open("archivo_nuevo.txt", "r")

archivo_leido.seek(10)   #Ubicamos el puntero en el caracter que se encuentra en la posición número 8

print(archivo_leido.read())

archivo_leido.close()
```

Si lo que queremos es que lea desde el comienzo hasta una posición determinada, entonces la ubicacion se la indicamos dentro de la función "read"

```python
#Utilizamos el comando open, pasándole dos parámetros: El nombre del archivo y el modo en el que lo queremos abrir 
archivo_leido = open("archivo_nuevo.txt", "r")

print(archivo_leido.read(7))
print(archivo_leido.read())
```

Regresamos el puntero a la posición 0 del archivo para leer todo 

```python
archivo_leido.seek(0) 

print(archivo_leido.read())

archivo_leido.close()
```

### <font color= #33691e> <i> Archivos de Tipo JSON

Los archivos de tipo JSON (JavaScript Object Notation) están formados por listas y diccionarios, de manera que son fáciles de comprender a simple vista 

Este tipo de archivo es muy utilizado en internet debido a que requiere de poco espacio, lo que facilita la transmisión de la información

<b> Ejemplo de Archivo JSON

```python
organigrama=[
  {
  "departamento":1,
  "nombre_depto":"Recursos Humanos",
  "director": "Laura Godinez",
  "empleados":[
    {
      "nombre":"Paul",
      "apellido":"Fierro"
    },{
      "nombre":"Jose",
      "apellido":"Bonilla"
    }]  
   },

  {
  "departamento":2,
  "nombre_depto":"Ingenierias",
  "director": "Salvador Medrano",
  "empleados":[
    {
      "nombre":"Carlos",
      "apellido":"Oropeza"
    },{
      "nombre":"Manuel",
      "apellido":"Medina"
    }, {
      "nombre":"Roberto",
      "apellido":"Aguirre"
    }]
  }   
]   
```

```python
organigrama
```

#### <b> <font color= #990000 > <i> Convirtiendo un objeto JSON a un archivo JSON 

```python
import json
json_string = json.dumps(organigrama)
```

```python
with open('organigrama.json', 'w') as file:
    file.write(json_string)
```

```python
# 2. Un vez creado, leelo y muestra la información en pantalla
leer_json = open("organigrama.json", "r")
informacion = leer_json.read()
informacion = json.loads(informacion)
informacion
```
