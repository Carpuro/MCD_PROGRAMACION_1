```python
#Librería para conectar Python con SQL Server
import pyodbc  #instalar en el prompt de anaconda con pip install pyodbc

#Librería para manejo y análisis de datos
import pandas as pd 

#Nombre del servidor donde está instalado SQL Server Express
#'Localhost\SQLEXPRESS' Indica que el motor está en esta misma computadora
server = r'localhost\SQLEXPRESS'

#Nos conectamos a la base de datos 'master'
#Siempre debemos conectarnos primero a una base existente 
#'master' es la base princial del sistema SQL Server
database='master'
    
try:                               
    
    #Creamos la conexión usando autentificación de Windows (Trusted_Connection)
    conexion = pyodbc.connect(driver='{SQL server};'
                              f'SERVER={server};'
                              f'DATABASE={database};'
                              'Trusted_Connection=yes;',
                              autocommit=True #Necesario para permitir CREATE DATABASE
    ) 
    
    print('Conexión exitosa')

except: 

    print('La conexión no fué exitosa')
    
#Creamos un cursor para enviar comandos SQL al servidor
cursor = conexion.cursor()

#Ejecutamos un comando SQL para crear una base de datos llamada DB_Python # Cada que se ejecuta, la base que se creo, la marca como que ya existe. 
cursor.execute("CREATE DATABASE DB_Python4")  #Este lo comenté porque ya había creado la base. 
print ("Base de datos creada:DB_Python4")


#Nos aseguramos de cerrar la conexión
conexion.close()
```

```python
import pyodbc
import pandas as pd

server = 'localhost\\SQLEXPRESS'
bd = 'DB_Python4'
usuario = 'Usuario1'
contrasena= '12345'

try:                               
    conexion = pyodbc.connect(driver='{SQL server}', host = server, database = bd, UID = usuario, PWD = contrasena)                              
    print('Conexión exitosa')
except: 
    print('La conexión no fué exitosa')

# CARGAR DATOS DEL CSV
ruta = r"C:\Users\leona\OneDrive\Desktop\MCD\1er_Semestre\Programacion\Sesion_11\MaratonNY.csv"
df = pd.read_csv(ruta)

# Convertir NaN → None (para que SQL acepte NULL)
df = df.where(pd.notnull(df), None)

cursor = conexion.cursor()

for index, row in df.iterrows():
    try:
        cursor.execute("""
            INSERT INTO MaratonNY (Corredor, place, gender, age, home, time)
            VALUES (?, ?, ?, ?, ?, ?)
        """,
        int(row['Corredor']),
        int(row['place']),
        row['gender'],
        int(row['age']),
        row['home'],      # ahora None → NULL
        float(row['time']))
        
    except Exception as e:
        print("\n❌ ERROR EN LA FILA:", index)
        print(row)
        print("Descripción del error:", e)
        break

conexion.commit()

maraton = pd.read_sql("SELECT * FROM MaratonNY", conexion)
df_maraton = pd.DataFrame(maraton)

conexion.close()

```
