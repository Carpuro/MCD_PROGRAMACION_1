### <b> <font color=  #6600CC > <i> Where, igual a, diferente a, Like y Not Like

```python
import pandas as pd
import pyodbc

# Leer el archivo Excel
ruta_excel = r"Datos_Personales_Python.xlsx"
Inf = pd.read_excel(ruta_excel, engine='openpyxl')

# Convertir todo a texto para evitar errores de tipo
Inf = Inf.fillna('').astype(str)

# Convertir DataFrame a lista de tuplas
tuplas_lista = [tuple(x) for x in Inf.values.tolist()]

# Configurar conexión
server = 'MONTSERRAT'
bd = 'DB_PYTHON'

try:
    conexion = pyodbc.connect(
        'DRIVER={ODBC Driver 17 for SQL Server};'
        f'SERVER={server};'
        f'DATABASE={bd};'
        'Trusted_Connection=yes;'
    )
    conexion.autocommit = True
    print('Conexión exitosa')

    cursor = conexion.cursor()

    # Crear la tabla si no existe
    cursor.execute("""
        IF OBJECT_ID('Datos_Personales_Python', 'U') IS NULL
        CREATE TABLE Datos_Personales_Python (
            Id_num INT PRIMARY KEY,
            Nombre NVARCHAR(100),
            Correo NVARCHAR(100),
            Direccion NVARCHAR(150),
            Ciudad NVARCHAR(100),
            Telefono NVARCHAR(20),
            Genero NVARCHAR(20),
            Estado_Civil NVARCHAR(20)
        )
    """)

    # Insertar o actualizar datos
    cursor.executemany("""
        MERGE INTO Datos_Personales_Python AS destino
        USING (SELECT ? AS Id_num, ? AS Nombre, ? AS Correo, ? AS Direccion, 
                      ? AS Ciudad, ? AS Telefono, ? AS Genero, ? AS Estado_Civil) AS origen
        ON destino.Id_num = origen.Id_num
        WHEN MATCHED THEN
            UPDATE SET 
                Nombre = origen.Nombre,
                Correo = origen.Correo,
                Direccion = origen.Direccion,
                Ciudad = origen.Ciudad,
                Telefono = origen.Telefono,
                Genero = origen.Genero,
                Estado_Civil = origen.Estado_Civil
        WHEN NOT MATCHED THEN
            INSERT VALUES (origen.Id_num, origen.Nombre, origen.Correo, origen.Direccion,
                           origen.Ciudad, origen.Telefono, origen.Genero, origen.Estado_Civil);
    """, tuplas_lista)

    conexion.commit()
    print('Datos cargados correctamente en SQL Server')

except Exception as e:
    print('Error:', e)

finally:
    if 'conexion' in locals():
        conexion.close()
        print('Conexión cerrada')

```

Pandas tiene dos funciones que permiten accesar más fácilmente a la información contenida en un Dataframe

Estas dos funciones son loc e iloc

Cuando conocemos los nombres de las filas y las columnas, podemos utilizar la función loc y cuando no los conocemos o queremos hacer referencia a las mismas por medio de números, utilizamos iloc (la "i" hace referencia a "integer")
