### <b> <font color=  #6600CC > <i> Funciones de Agrupamiento (sum, count, min, max y average)

```python

server = 'PROBOOK-ROCIO'
bd = 'DB_Python2'
    
try:                               
    
    conexion = pyodbc.connect(driver='{SQL server}', host = server, database = bd)                              
    print('Conexión exitosa')

except: 

    print('La conexión no fué exitosa')
    
    
suma = pd.read_sql("SELECT SUM(time)FROM MaratonNY", conexion)

cantidades = pd.read_sql("SELECT COUNT(time)FROM MaratonNY", conexion)

minimo = pd.read_sql("SELECT MIN(time)FROM MaratonNY", conexion)

maximo = pd.read_sql("SELECT MAX(time)FROM MaratonNY", conexion)

promedio = pd.read_sql("SELECT AVG(time)FROM MaratonNY", conexion)

cantidad_paises = pd.read_sql("SELECT COUNT(DISTINCT(home)) FROM MaratonNY", conexion)

#Nos aseguramos de cerrar la conexión
conexion.close()
```

```python
suma
```

```python
type(suma)
```

```python
print(f'La suma de los tiempos es {suma.iloc[0][0]}')
print(f'La cantidad de tiempos registrados es {cantidades.iloc[0][0]}')
print(f'El tiempo mínimo registrado es {minimo.iloc[0][0]}')
print(f'El tiempo máximo registrado es {maximo.iloc[0][0]}')
print(f'El tiempo promedio registrado es {promedio.iloc[0][0]}')
print(f'En el maratón participaron {cantidad_paises.iloc[0][0]} paises')
```

<b> <font color=  #FF00FF > <i> Accesando a los valores de un dataframe con loc e iloc

Pandas tiene dos funciones que permiten accesar más fácilmente a la información contenida en un Dataframe

Estas dos funciones son loc e iloc

Cuando conocemos los nombres de las filas y las columnas, podemos utilizar la función loc y cuando no los conocemos o queremos hacer referencia a las mismas por medio de números, utilizamos iloc (la "i" hace referencia a "integer")



<b> <font color=  #FF00FF > <i> Ejemplos con iloc

```python
carreras
```

```python
type(carreras)
```

```python
carreras[1,2] #Aún cuando "carreras" es un dataframe, este código nos arrojará un error 
```

```python
carreras.iloc[1][2] #Para obtener la información que se encuentra en la fila 1, columna 2, utilizamos iloc
```

```python
carreras.iloc[1,2] #Para obtener la información que se encuentra en la fila 1, columna 2, utilizamos iloc
```

```python
carreras.loc[1]["gender"] #Aquí utilizamos el número uno para la fila debido a que las filas no tienen un nombre
```

```python
carreras.loc[1,"gender"] #Aquí utilizamos el número uno para la fila debido a que las filas no tienen un nombre
```

```python
carreras.iloc[1] #Para obtener la información que se encuentra en la fila 1
```

```python
carreras.iloc[2:5] #Trae las filas 2 a 4 
```

```python
carreras.iloc[2:5, [2,3]] #Trae las filas 2 a 4, pero solamente las columnas gender y age 
```

```python
carreras.iloc[2:5, 2:5] #Trae las filas 2 a 4 y las columnas de gender a age 
                        #Cuando se le pide un rango de valores no se utilizan corchetes
                        #Los rangos toman en cuenta un número menos en la cifra que se encuentra a la derecha de los dos puntos
```

```python
carreras.iloc[[2,5], [2,5]] #Trae las filas 2 y 5, y las columnas gender y time (la segunda y la quinta)
```

```python
carreras.set_index("Corredor", inplace=True) #Con set_index le indicamos que la columna Corredor contiene los nombres de las filas
                                             #Con inplace=True le indicamos que el cambio se quedará guardado en el dataframe Carreras
```

```python
carreras.loc[13594,"gender"] #Ahora que ya las filas tienen un nombre, podemos pasarle el número de corredor en las filas
```

```python
carreras.loc[13594,["gender", "age", "home","time"]] 
```

```python
carreras.loc[13594,"gender":"time"] 
```

```python
carreras.loc[13594:9476, "gender":"time"] 
```

```python
carreras.iloc[1:5, 1:5] #En las columnas, ahora gender ya es la número 1 debido a que Corredor ya es un indice y no una columna
```
